```python
import pandas as pd
```

```python
data = pd.read_json('../../skaters.ndjson', lines=True)
```

```python
# columns useful to talk about a particular athlete
useful_columns = ['skaters_id', 'first_name', 'last_name', 'gender', 'nationality_code', 'organization_code', 'thumbnail_image', 'status', 'created_at', 'updated_at', 'discipline', 'is_favourite', 'results', 'details']

filtred_data = data[useful_columns]

with open('filtred_skaters.ndjson', 'w', encoding='utf-8') as f:
    f.write(filtred_data.to_json(orient='records', lines=True))
```
